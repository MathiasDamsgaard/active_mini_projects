```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(index=['A','B','C','D'])
dfl = {}
for name in os.listdir('./Samples/'):
    print(name)
    data = pd.read_csv(f'./Samples/{name}', index_col=0)
    #df[name.replace(".csv","")] = data.T.values.tolist()
    dfl[name.replace(".csv","")] = data
```

```python
def print_info(data):
    print(f'Num datapoinst: {len(data)}')
    for name,(a,b) in zip(data.columns, zip(data.mean(), data.var())):
        print(f'{name}: mean {a:.5f}     var {b:.5f}')
```
